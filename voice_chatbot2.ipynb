```python
import streamlit as st
import speechrecognition as sr
import openai
import os
```

```python
st.title("🤖 شات بوت بالصوت والكتابة - عربي")
```

```python
# اختيار طريقة الإدخال
mode = st.radio("اختر طريقة الإدخال:", ["كتابة", "صوت"])
```

```python
if mode == "كتابة":
    user_input = st.text_input("✍️ اكتب سؤالك هنا")
else:
    if st.button("🎙️ سجل صوت"):
        recognizer = sr.Recognizer()
        with sr.Microphone() as source:
            st.write("🎤 اتكلم دلوقتي...")
            audio = recognizer.listen(source)
            try:
                user_input = recognizer.recognize_google(audio, language="ar-AR")
                st.success(f"النص المستخرج: {user_input}")
            except Exception as e:
                st.error(f"حصل خطأ: {e}")
                user_input = ""
```

```python
# الرد من OpenRouter
if st.button("🚀 ابعت"):
    if user_input.strip() != "":
        with st.spinner("🤔 بيرد..."):
            response = openai.ChatCompletion.create(
                model="openchat/openchat-7b",  # ممكن تغيريه لأي موديل متاح على OpenRouter
                messages=[{"role": "user", "content": user_input}],
            )
            bot_reply = response.choices[0].message["content"]
            st.success(f"🤖 الرد: {bot_reply}")

            # تحميل الرد كملف
            st.download_button("📂 حمل الرد كملف", bot_reply, file_name="reply.txt")
    else:
        st.warning("من فضلك اكتب أو سجل سؤال الأول.")
```
